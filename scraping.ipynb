# Extraccion de los datos mediante tecnicas de Scraping

```python
"""
    Librerias utilizadas
"""
import re
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient
from threading import Thread
```

```python
# Conexion a MongoDB
client = MongoClient('localhost', 27017)
db = client['dataframe']
coleccion_enlaces = db['estadisticas']
```

```python
"""
Aqui indico el indice del nombre de los mapas, 
como una tupla con los nombres de los agentes que hay en el videojuego
"""
# 1. Bind, 2. Haven, 3. Split, 5. Ascent, 6. IceBox, 8. Breeze, 9. fracture, 10. Pearl, 11. Lotus

agentes = ('jett','astra','breach','brimstone','chamber','cypher','fade','gekko','harbor','kayo','killjoy','neon','omen','phoenix','raze', 'reyna', 'sage', 'skye','sova','viper','yoru')
```

```python
"""Insercion de los datos scrapeados a Mongo
"""
def insert_mondodb(header_valores, fila_valores,mapa_id):
    data_frame = {
        "Mapa": mapa_id,
        header_valores[0].text: fila_valores[0],
        header_valores[1].text: fila_valores[1],
        header_valores[2].text: fila_valores[2],
        header_valores[3].text: fila_valores[3],
        header_valores[4].text: fila_valores[4],
        header_valores[5].text: fila_valores[5],
        header_valores[6].text: fila_valores[6],
        header_valores[7].text: fila_valores[7],
        header_valores[8].text: fila_valores[8],
        header_valores[9].text: fila_valores[9],
        header_valores[10].text: fila_valores[10],
        header_valores[11].text: fila_valores[11],
        header_valores[12].text: fila_valores[12],
        header_valores[13].text: fila_valores[13],
        header_valores[14].text: fila_valores[14],
        header_valores[15].text: fila_valores[15],
        header_valores[16].text: fila_valores[16],
        header_valores[17].text: fila_valores[17],
        header_valores[18].text: fila_valores[18],
        header_valores[19].text: fila_valores[19],
        header_valores[20].text: fila_valores[20],
    }   

    operacion = coleccion_enlaces.insert_one(data_frame)
```

```python
"""Hilo para la extracion de informacion de los agentes en el mapa Lotus
"""
def extraction_lotus(agente, header_valores, mapa_id=11):
    url = f'https://www.vlr.gg/stats/?event_group_id=all&event_id=all&region=all&country=all&min_rounds=50&min_rating=1500&agent={agente}&map_id={mapa_id}&timespan=all'
    request = requests.get(url)
    soup = BeautifulSoup(request.text)
    tabla = soup.find('table')
    body_tabla = tabla.find('tbody')
    filas_tabla = body_tabla.find_all('tr')
    for fila in filas_tabla:
        valores_fila = fila.find_all('td')
        fila_valores = []
        for valor in valores_fila:
            fila_valores.append(valor.text.strip())

        fila_valores[1] = agente
        
        insert_mondodb(header_valores, fila_valores, mapa_id)
```

```python
# Extraer los valores de la cabezera de la tabla
url_cabezera = 'https://www.vlr.gg/stats/?event_group_id=43&event_id=1555&series_id=all&region=sa&country=all&min_rounds=200&min_rating=1550&agent=all&map_id=all&timespan=30d'
request = requests.get(url_cabezera)
soup = BeautifulSoup(request.text)
tabla = soup.find('table')
cabezera_tabla = tabla.find('thead')
header_valores = cabezera_tabla.find_all('th')

#Extrae los datos de la tabla a traves de hilos

try:
    for agente in agentes:

        extraction_bind_t = Thread(target=extraction_bind(agente,header_valores))

        extraction_ascent_t = Thread(target=extraction_ascent(agente,header_valores))

        extraction_fracture_t = Thread(target=extraction_fracture(agente,header_valores))

        extraction_breeze_t = Thread(target=extraction_breeze(agente,header_valores))

        extraction_haven_t = Thread(target=extraction_haven(agente,header_valores))

        extraction_icebox_t = Thread(target=extraction_icebox(agente,header_valores))

        extraction_split_t = Thread(target=extraction_split(agente,header_valores))
        
        extraction_pearl_t = Thread(target=extraction_pearl(agente,header_valores))

        extraction_lotus_t = Thread(target=extraction_lotus(agente,header_valores))

        extraction_bind_t.start()
        extraction_ascent_t.start()
        extraction_fracture_t.start()
        extraction_breeze_t.start()
        extraction_haven_t.start()
        extraction_icebox_t.start()
        extraction_split_t.start()
        extraction_pearl_t.start()
        extraction_lotus_t.start()

    print("Ciclo Finalizado")

except Exception as e:
    print("Error en hilo")
    print(e)
```
